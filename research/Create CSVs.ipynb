```python
import pandas as pd
import os
```

```python
path= '../data/'
files = [path + file for file in os.listdir(path) if file.endswith(('.xlsx'))]

files
```

```python
def save_df(df: pd.DataFrame, name: str):
    df.to_csv(
        path + name.split('/')[-1].split('.')[0] + '.csv',
        index=False
    )
```

```python
dataframes = dict()
```

```python
dataframes['cococola'] = pd.read_excel(files[0], sheet_name=0)
dataframes['cococola'].tail()
```

```python
save_df(df=dataframes['cococola'], name=files[0])
```
